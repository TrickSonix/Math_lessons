Задача 1

О непрерывной равномерно распределённой случайной величине B известно, что её дисперсия равна 0.2. Можно ли найти правую границу величины B и её среднее значение зная, что левая граница равна 0.5? Если да, найдите их.

Для равномерно распределенной случайной величины справедливо: f(x) = 1/(b-a) при a<x<b а дисперсия D(x) = (b-a)^2/12, таким образом можно найти правую границу:
(b-0.5)^2/12=0.2 =>b^2-b-2.15=0 => b~2.05
Отсюда можно найти среднее значение: M(x) = 1.275

Задача 2

Коробки с шоколадом упаковываются автоматически. Их средняя масса равна 1.06 кг. Известно, что 5% коробок имеют массу, меньшую 1 кг. Найдите: а) среднее квадратическое отклонение массы коробки, б) процент коробок, имеющих массу больше 1.1 кг.

В данном случае имеем дело с нормальным распределением. M(x) = 1.06. Вероятность того, что коробка будет иметь массу меньшую чем 1 кг вычисляется с помощью функции Лапласа, откуда мы можем найти среднеквадратичное отклонение:
P(x<1)=Ф(-0.06/σ)-Ф(-∞)=0.05
С помощью таблицы найдем значения для функции лапласа:
Ф(-∞)=-0.5 => Ф(0.06/σ)=0.45
В таблице находим какому значению аргумента соответствует данное значение функции и получаем:
0.06/σ=0.6 => σ=0.1
Теперь, когда мы знаем функцию плотности распределения вероятностей, можно посчитать вероятность того, что коробка будет весить больше 1.1 кг:
P(x>1.1) = Ф(∞)-Ф((1.1-1.06)/0.1)=0.5-Ф(0.4)=0.5-0.311=0.189. Ответ: 18.9%

Задача 3

Коробка содержит 30 конфет. Известно, что масса каждой конфеты распределена равномерно в промежутке от 12 до 14 граммов. Используя центральную предельную теорему, найти вероятность, что масса всей коробки будет: а) меньше 390 граммов, б) больше 395 граммов, в) от 380 до 400 граммов.

Массой самой коробки можно пренебречь.

Пусть Х - масса конфеты. Тогда M(X) = (14+12)/2=13, σ^2 = (14-12)^2/12=1/3. По центральной предельной теореме масса 30 конфет будет иметь нормальное распределение с M(X) = 30*13=390 и σ^2 = 30*1/3=10. Отсюда можно найти необходимые вероятности:
а)P(x<390) = Ф(0)-Ф(-∞)=0.5 б)P(x>395) = Ф(∞)-Ф(1.58)=0.5-0.443=0.057 в)P(380<x<400) = Ф(3.16)+Ф(3.16) = 0.499+0.499=0.998
