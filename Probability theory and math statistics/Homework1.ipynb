Задача 1

Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

Число исходов при котором был вытянут хотя бы один туз = число исходов, при котором был вытянут один туз + число исходов при котором было вытянуто 2 туза + число исходов при котором было вытянуто 3 туза + число исходов при котором было вытянуто 4 туза.

В данном случае нам не важен порядок, поэтому нас интересует число сочетаний. Вытянут один туз: число сочетаний из 4 по 1 умножить на число сочетаний из 48 по 3.

И так далее для всех остальных случаев.

Для 4 тузов у нас есть только 1 исход.

```python
def combination(n, k):
    return np.math.factorial(n) // (np.math.factorial(k)*np.math.factorial(n-k))
```

```python
combination(4,1)*combination(48, 3) + combination(4,2)*combination(48, 2) + combination(4,3)*combination(48, 1) + 1
```

Задача 2

Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что рядом будут сидеть: а) два определённых человека? б) три определённых человека?

а)Пусть событие А - два определенных человека сидят рядом. P(A) = m/n, где m - кол-во благоприятных исходов для А, n - кол-во всех исходов. В данном случае n = 7!. Благоприятные исходы для А, такие, где два определенных человека сидят вместе. Всего на 7 мест есть 6 пар мест находящихся рядом и на каждую пару мест, два определенных человека могут сесть 2! способами. И оставшиеся 5 человек 5! способами.
б)Рассуждения те же. Благоприятные исходы для А, такие, где три определенных человека сидят вместе. Всего на 7 мест есть 5 различных рядом находящихся мест для 3 человек и на каждое такое место 3 определенных человека могут сесть 3! способами. И оставшиеся 4 человека 4! способами.

```python
2*6*np.math.factorial(5)/np.math.factorial(7)
```

```python
5*6*np.math.factorial(4)/np.math.factorial(7)
```

Задача 3

Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Какова вероятность того, что среди трёх наугад выбранных вопросов студент знает: а) все? б) два?

а)Вероятность того, что студент знает ответ на первый вытянутый вопрос 50/60. Далее мы тянем следующий вопрос из оставшихся 59. Вероятность что студент знает на него ответ 49/59. Далее третий и его вероятность 48/58. Поскольку знание ответа на эти вопросы - независимые события, получаем вероятность того, что студент знает все три вопроса: (50/60)*(49/59)*(48/58). Так же можно получить тот же ответ другим рассуждением: Всего 3 вопроса из 60 можно вытащить 60!/(3!*(60-3)!)=60*59*58/6 способами, кол-во же способов вытащить билет, ответ на который знает студент 50!/(3!*(50-3)!)=50*49*48/6, таким образом вероятность что студент знает ответ на все 3 вопроса это (50*49*48)/(60*59*58).

б)Всего 3 вопроса из 60 можно вытащить 60!/(3!*(60-3)!)=60*59*58/6 способами, а вытянуть 2 вопроса, на которые студент знает ответ и 1, на который не знает можно 50!/(2!*(50-2)!) * 10!/9!=50*49*10/2. Таким образом вероятность такого события (50*49*10*6)/(60*59*58*2).

```python
(50*49*48)/(60*59*58)
```

```python
(50*49*10*6)/(60*59*58*2)
```

Задача 4

Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?

Если события независимые, то Р(А|B) = P(A).
P(A) = 1/2
P(B) = 1/2
P(AB) - выпадение 4 или 6 = 2/6
P(A|B) = P(AB)/P(B) = (2/6)/(1/2) = 2/3 => События зависимые.

Задача 5

Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.

Вопрос: какова вероятность, что вы действительно больны ей?

Пусть А - заболеть болезнью. B - тест для нас положительный.
Тогда P(A|B) = P(A)*P(B|A)/P(B). P(A) = 0.001, P(B|A) = 0.99. P(B) = P(B|A)*P(A) + P(B|-A)*P(-A) = 0.99*0.001 + 0.01*0.999 = 0.01098. Тогда P(A|B) = 0.001*0.99/0.01098

```python
0.99*0.001 + 0.01*0.999
```

```python
(0.001*0.99)/0.01098
```
